**Breast Cancer Classification using Neural Network**

Importing the Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

Loading the data

```python
# loading the data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

**The 'data' key contains all the feature values and the 'target' key contains all the labels**

```python
# loading the data to a data frame using pandas
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
data_frame.head()
```

```python
# adding the 'target' column to the data frame
data_frame['label'] = breast_cancer_dataset.target
```

```python
data_frame.head()
```

Hence, a new 'target' column has been added 

```python
data_frame.tail()
```

```python
data_frame.shape
```

```python
# checking for non-null values
data_frame.info()
```

```python
# getting the statistical measures 
data_frame.describe()
```

```python
# finding how many tumors are benign or malignant in the dataset (0-malignant)
data_frame['label'].value_counts()
```

```python
#finding the mean of the measures of the benign and malignant tumors
data_frame.groupby('label').mean()
```

Splitting the features and the target

```python
X = data_frame.drop(columns='label', axis=1)
Y = data_frame['label']
```

```python
print(X)
```

```python
print(Y)
```

Serparating the training and testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Performing Data Standardization

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

X_train_std = scaler.fit_transform(X_train)

X_test_std = scaler.transform(X_test)
```

**Building the Neural Network**

```python
# importing tensorflow and Keras
import tensorflow as tf 
tf.random.set_seed(3) #
from tensorflow import keras
```

```python
# setting up the layers of Neural Network
#using sequential model
model = keras.Sequential([
                          keras.layers.Flatten(input_shape=(30,)),
                          keras.layers.Dense(20, activation='relu'),
                          keras.layers.Dense(2, activation='sigmoid')
])
```

```python
# compiling the NN

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# training the Neural Network

history = model.fit(X_train_std, Y_train, validation_split=0.1, epochs=12)
```

Plotting the accuracy 

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'upper right')
```

Finding the accuracy of the model on test data

```python
loss, accuracy = model.evaluate(X_test_std, Y_test)
print(accuracy)
```
